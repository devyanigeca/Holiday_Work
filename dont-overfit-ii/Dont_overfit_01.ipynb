### applying logistic regression 

```python
"""
    Logistic function = sigmoid 
    i.e 1//1+exp(-x)
    hence our eq is H = 1/1+exp(-(mx+c)) unlike linear regg where H = mx +c 
    
    cost func : -1/N * Sum( yi(logH) + (1-yi)(1-logH) )
    
"""
y=[]
def findit(m,c,lr,ip,tr):
    ctheta = 0
    mtheta = 0
    for i in range(len(ip)):
        ctheta = ctheta + -((-int(ip[i])+(1-int(ip[i])))*(1+(math.exp(-(m*int(ip[i])+c))))*(math.exp(-(m*int(ip[i])+c))))/len(ip)
        mtheta = mtheta + -((int(ip[i])-(1-int(ip[i])))*(1+(math.exp(-(m*int(ip[i])+c))))*(math.exp(-(m*int(ip[i])+c)))*(int(ip[i])))/len(ip)
    c = c - lr * ctheta
    m = m - lr * mtheta
    print(c," ",m)
    y.append(c)
    return (c,m)

def logfunc(ip,tr):
    m = 0
    c = 0
    lr = 0.00005
    for i in range(1000):
        print(i)
        cnew ,mnew = findit(m,c,lr,ip,tr)
        c = cnew
        m = mnew
    print('c is : ',c,' m is : ',m)    
    
```

```python
logfunc(df['0'],df['target'])
```
